Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest_score = max(employees.values())

top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

if top_performers:
    names_str = ", ".join(top_performers)
    output = f"Top Performers Eligible for Bonus: {names_str} (Score: {highest_score})"
    print(output)
else:
    print("No employees found.")
```

Problem Statement 2: Search Query Keyword Analysis

```python
search_query = "Buy mobile phone buy phone online"
```

```python
import collections
import re

cleaned_query = search_query.lower()

cleaned_query = re.sub(r'[^a-z0-9\s]', '', cleaned_query)

words = cleaned_query.split()
word_counts = collections.Counter(words)

output_keywords = {
    word: count for word, count in word_counts.items() if count > 1
}

if output_keywords:
    print("Keywords searched more than once:")
    for word, count in output_keywords.items():
        print(f"- {word}: {count}")
else:
    print("No keywords found more than once.")
```

Problem Statement 3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
```

```python
valid_readings = []

for hour_index, reading_value in enumerate(sensor_readings):
    if reading_value % 2 == 0:
        valid_readings.append((hour_index, reading_value))

if valid_readings:
    print("Valid Sensor Readings (hour_index, reading_value):")
    for hour, value in valid_readings:
        print(f"- Hour {hour}: {value}")
else:
    print("No valid sensor readings found.")
```

Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
```

```python
import collections

domains = []
for email in emails:
    domain = email.split('@')[1]
    domains.append(domain)

domain_counts = collections.Counter(domains)

total_emails = len(emails)

if total_emails > 0:
    for domain, count in domain_counts.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")
else:
    print("No emails provided for analysis.")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
```

```python
if sales:
    average_sales = sum(sales) / len(sales)
else:
    average_sales = 0

spike_threshold = average_sales * 1.30

found_spikes = False
for day_number, sale_value in enumerate(sales, 1):
    if sale_value > spike_threshold:
        print(f"Day {day_number}: {sale_value}")
        found_spikes = True
```

Problem Statement 6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
```

```python
import collections

id_counts = collections.Counter(user_ids)

duplicates_found = False
for user_id, count in id_counts.items():
    if count > 1:
        print(f"{user_id} → {count} times")
        duplicates_found = True

if not duplicates_found:
    print("No duplicate user IDs found.")
```
