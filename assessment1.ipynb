1. User Login Check

```python
def login(username, password):
    if username == "admin" and password == "1234":
      print("Login Successful")
    else:
      print("Invalid Credentials")

username=input("Enter the username: ")
password=input("Enter the password: ")
login(username, password)
```

2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
    pass_count+=1
  else:
    fail_count+=1
print("total number of pass students:",pass_count)
print("total number of fail students:",fail_count)
```

3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
names = [name.strip().lower() for name in names]
print(names)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}', Length: {length}")
    if length > 10:
        print("  -> Long message!")
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0
for entry in logs:
    if entry == "ERROR":
        error_count += 1

print(f"Number of ERROR messages: {error_count}")
```
