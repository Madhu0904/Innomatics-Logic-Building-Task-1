

Problem Statement 1:Smart Parking Lot Management System




```python
def manage_parking_lot(capacity, vehicle_logs):
    current_parked_vehicles = 0
    peak_parked_vehicles = 0
    alerts = []

    for log in vehicle_logs:
        if log == "IN":
            current_parked_vehicles += 1
            if current_parked_vehicles > capacity:
                alerts.append(f"ALERT: Parking capacity exceeded! Currently {current_parked_vehicles} vehicles, capacity {capacity}.")
        elif log == "OUT":
            if current_parked_vehicles > 0:
                current_parked_vehicles -= 1
            else:
                alerts.append("WARNING: Attempted vehicle exit, but no vehicles are currently parked.")

        # Update peak parking usage
        peak_parked_vehicles = max(peak_parked_vehicles, current_parked_vehicles)

    parking_status = "Full" if current_parked_vehicles >= capacity else "Available"

    print(f"Currently Parked Vehicles: {current_parked_vehicles}")
    print(f"Parking Status: {parking_status}")

    if alerts:
        print("\n--- Alerts ---")
        for alert in alerts:
            print(alert)
```

```python
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking_lot(parking_capacity, vehicle_logs)
```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance_km, traffic_level, weather):
    base_time_minutes = distance_km * 5

    total_delay_minutes = 0

    #traffic delays
    if traffic_level.lower() == "high":
        total_delay_minutes += 10  # add 10 minutes
    elif traffic_level.lower() == "medium":
        total_delay_minutes += 5   # add 5 minutes

    #weather delays
    if weather.lower() == "rainy":
        total_delay_minutes += 5   # add 5 minutes
    elif weather.lower() == "snowy":
        total_delay_minutes += 10  # add 10 minutes

    final_eta_minutes = base_time_minutes + total_delay_minutes

    print(f"Estimated Delivery Time: {final_eta_minutes} minutes")

distance = 8 # km
traffic = "High"
weather = "Rainy"

print("test case 1:")
estimate_delivery_time(distance, traffic, weather)

print("\nTest Case 2")
estimate_delivery_time(15, "Medium", "Clear")
```

 Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def analyze_seat_occupancy(total_seats, booked_seats_list):
    num_booked_seats = sum(booked_seats_list)

    if total_seats <= 0:
        print("Error: Total seats must be a positive number.")
        return

    occupancy_percentage = (num_booked_seats / total_seats) * 100

    show_status = "Available"
    suggestions = []

    if occupancy_percentage == 100:
        show_status = "Housefull"
        suggestions.append("Suggest opening additional shows due to high demand.")
    elif occupancy_percentage >= 75:
        show_status = "Almost Full"
        suggestions.append("Consider opening additional shows soon as demand is high.")
    elif occupancy_percentage >= 50:
        show_status = "Moderate Occupancy"
    else:
        show_status = "Low Occupancy"

    print(f"Occupancy: {occupancy_percentage:.0f}%")
    print(f"Show Status: {show_status}")

    if suggestions:
        print("--- Suggestions ---")
        for suggestion in suggestions:
            print(suggestion)

total_seats_1 = 200
booked_seats_1 = [1] * 150  # 150 booked seats

analyze_seat_occupancy(total_seats_1, booked_seats_1)

```

 Problem Statement 4:Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):
    if not cpu_readings:
        print("Error: No CPU readings provided.")
        return

    average_cpu = sum(cpu_readings) / len(cpu_readings)
    server_status = "Unknown"

    if average_cpu < 50:
        server_status = "Normal"
    elif 50 <= average_cpu < 80:
        server_status = "Warning"
    elif average_cpu >= 80:
        server_status = "Critical"

    print(f"Average CPU Load: {average_cpu:.0f}%")
    print(f"Server Status: {server_status}")


cpu_readings_1 = [45, 60, 70, 85, 90]

print("Processing Sample Input:")
classify_server_load(cpu_readings_1)
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def monitor_resource_usage(resource_usage_hours):
    #overuse thresholds
    overuse_thresholds = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused_resources = []
    energy_alert = "No"

    for resource, usage in resource_usage_hours.items():
        if resource in overuse_thresholds and usage > overuse_thresholds[resource]:
            overused_resources.append(resource)
            energy_alert = "Yes"

    if overused_resources:
        print(f"Overused Resources: {', '.join(overused_resources)}")
    else:
        print("Overused Resources: None")
    print(f"Energy Alert: {energy_alert}")
```

```python

resource_usage_1 = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_usage_1)
```

 Problem Statement 6: Online Event Registration Capacity Controller


```python
def manage_event_registrations(event_capacity, num_registrations_attempted):
    confirmed_registrations = min(num_registrations_attempted, event_capacity)
    waitlisted_users = max(0, num_registrations_attempted - event_capacity)

    registration_status = "Open"
    if num_registrations_attempted >= event_capacity:
        registration_status = "Closed"

    print(f"Confirmed Registrations: {confirmed_registrations}")
    print(f"Waitlisted Users: {waitlisted_users}")
    print(f"Registration Status: {registration_status}")


event_capacity_1 = 100
num_registrations_attempted_1 = 105
manage_event_registrations(event_capacity_1, num_registrations_attempted_1)
```
