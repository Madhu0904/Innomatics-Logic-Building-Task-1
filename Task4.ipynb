Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and plan in valid_plans:
        return True
    return False

while True:
    amount = int(input("Enter amount: "))
    plan = int(input("Enter plan: "))

    if validate_recharge(amount, plan):
        print("Valid recharge")
        break
    else:
        print("Invalid, try again")
```

Problem 2: Inventory Reorder Alert System


```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"Product: {product}, Stock: {stock} - Reorder Alert!")
        else:
            print(f"Product: {product}, Stock: {stock} - Stock OK")

products_data = {
    "Sweaters": 20,
    "T-shirts": 10,
    "Shirts": 5,
    "Jackets": 25,
    "Denim": 12,
    "Belts": 30
}

print("\n--- Inventory Status ---")
check_inventory(products_data)
```

 Problem 3: Student Result Processing System


```python
def process_student_results(marks):
    if not marks:
        print("No marks: 0")
        return

    total_marks = sum(marks)
    average_marks = total_marks / len(marks)

    print(f"Student Marks: {marks}")
    print(f"Average Marks: {average_marks:.2f}")

    if average_marks >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")

# testing cases:
print("--- Student 1 ---")
student1_marks = [70, 65, 80, 45, 75]
process_student_results(student1_marks)

print("\n--- Student 2 ---")
student2_marks = [30, 40, 48, 55, 25]
process_student_results(student2_marks)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_cab_fare(distance_km, is_peak_hour):
    base_fare = 50
    fare_per_km = 12

    total_fare = base_fare + (distance_km * fare_per_km)

    if is_peak_hour:
        total_fare *= 1.25  # 25% extra for peak hour

    return total_fare

while True:
    distance = float(input("Enter distance traveled in km: "))
    peak_hour_input = input("Is it a peak hour? (yes/no): ").lower()
    is_peak_hour = (peak_hour_input == 'yes')

    if distance < 0:
        print("Distance cannot be negative. Please try again.\n")
        continue

    fare = calculate_cab_fare(distance, is_peak_hour)
    print(f"Calculated Cab Fare: ₹{fare:.2f}")
    break # Exit loop
```

Problem 5: Employee Attendance Eligibility Checker


```python
def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        return "No attendance data"

    present_days = attendance_list.count('P')
    total_days = len(attendance_list)
    attendance_percentage = (present_days / total_days) * 100

    print(f"Attendance: {attendance_list}")
    print(f"Attendance Percentage: {attendance_percentage:.2f}%")

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

print("--- Employee 1 ---")
employee1_attendance = ['P', 'P', 'A', 'P', 'P', 'P', 'A', 'P', 'P', 'P']
print("Eligibility:", check_attendance_eligibility(employee1_attendance))

print("\n--- Employee 2 ---")
employee2_attendance = ['P', 'A', 'A', 'P', 'A', 'P', 'A', 'P', 'A', 'A']
print("Eligibility:", check_attendance_eligibility(employee2_attendance))
```

Problem 6: Password Strength Checker


```python
def check_password_strength(password):
    min_length = 8
    has_digit = False
    has_special_char = False
    special_chars = "@#$"

    # Minimum length
    if len(password) < min_length:
        return "Weak: Password must be at least 8 characters long."

    for char in password:
        # At least one digit
        if char.isdigit():
            has_digit = True
        # At least one special character
        if char in special_chars:
            has_special_char = True

    if not has_digit:
        return "Weak: Password must contain at least one digit."
    if not has_special_char:
        return "Weak: Password must contain at least one special character (@#$)."

    return "Strong: Password meets all criteria."

print("--- Password Strength ---")
print("Password 'myPass123!':", check_password_strength('myPass123!')) # Strong
print("Password 'password':", check_password_strength('password')) # Weak: no digit, no special
print("Password 'mypas12':", check_password_strength('mypas12')) # Weak: too short
print("Password 'mypassword123':", check_password_strength('mypassword123')) # Weak: no special
```
