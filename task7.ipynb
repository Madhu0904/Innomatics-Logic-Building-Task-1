**Problem 1: Smart Light Controller**


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")

    def display_status(self):
        print(f"{self.name} is {self.status}")

bedroom_light = SmartLight("Bedroom Light")
bedroom_light.turn_on()
```

Problem 2: Employee ID Card System

```python
class EmployeeIDCard:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

employee1 = EmployeeIDCard("Rahul", "EMP102", "IT")
employee1.display_id_card()
```

Problem 3: Mobile Contact Record


```python
class MobileContact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        print("Contact Saved")

    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

anita = MobileContact("Anita", "9876543210")
anita.display_contact()
```

Problem 4: Product Price Tag Generator


```python
class ProductPriceTag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

headphones = ProductPriceTag("Headphones", 2499)
headphones.print_tag()
```

Problem 5: Movie Rating Display System


```python
class MovieRating:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

inception = MovieRating("Inception", 4.8)
inception.display_movie_details()
```

Problem 6: Delivery Address Manager


```python
class DeliveryAddressManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

suman_delivery = DeliveryAddressManager("Suman", "Hyderabad")
suman_delivery.print_delivery_details()
```
