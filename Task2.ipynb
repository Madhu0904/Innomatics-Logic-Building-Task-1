Problem 1: Unique Words in a Sentence

```python
sentence = "python is easy and python is powerful"
words = sentence.lower().replace('.', '').split()
unique_words = set(words)

print(f"Count of unique words: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

Problem 2: Highest Salary from Employee Data

```python
employee_salaries = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
}
```

```python
highest_salary = 0
highest_paid_employee = ""

for employee, salary in employee_salaries.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_paid_employee = employee

print(f"Highest Salary: {highest_paid_employee}-{highest_salary}.")
```

Problem 3: Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]
if not numbers:
    print("empty list")
else:
    max_value = numbers[0]
    min_value = numbers[0]

    for number in numbers:
        if number > max_value:
            max_value = number
        if number < min_value:
            min_value = number

    print("List:", numbers)
    print(f"Max: {max_value}")
    print(f"Min: {min_value}")
```

Problem 4: Count Products Above a Price Threshold

```python
product_prices = [450, 1200, 899, 1500, 300]
price_threshold = 1000
products_above_threshold = 0

for price in product_prices:
    if price > price_threshold:
        products_above_threshold += 1

print(f"Products above {price_threshold}: {products_above_threshold}")
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
total_days = len(attendance)
present_days = attendance.count("P")

if total_days > 0:
    attendance_percentage = (present_days / total_days) * 100
    print(f"Attendance percentage: {attendance_percentage:.2f}%")
else:
    print("Attendance list is empty. Cannot calculate percentage.")
```

Problem 7: Count Character Frequency

```python
input_string = "hello world"
char_frequency = {}

for char in input_string:
    if char.isalnum():
        char_frequency[char] = char_frequency.get(char, 0) + 1

print(f"Character frequencies in '{input_string}':")
for char, count in char_frequency.items():
    print(f"'{char}': {count}")
```

Problem 9: Check if a Key Exists in a Dictionary

```python
employee_data = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "employee": True
}
key_to_check = "employee"

if key_to_check in employee_data:
    print(f"The key '{key_to_check}' exists in the dictionary.")
else:
    print(f"The key '{key_to_check}' does not exist in the dictionary.")
```

Problem 10: Calculate Average Marks

```python
marks = [85, 70, 90, 65, 80, 75]
if not marks:
    print("The list of marks is empty. Cannot calculate average.")
else:
    total_marks = sum(marks)
    number_of_subjects = len(marks)
    average_marks = total_marks / number_of_subjects

    print(f"Average Marks: {average_marks:.2f}")
```
