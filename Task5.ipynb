Problem 1: Social Media – Post Engagement Analyzer


```python
def social_media(likes):

    total_likes = 0
    for like_count in likes:
        total_likes += like_count

    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"

    return total_likes, post_status

#Interactions:
post_likes = [200, 350, 500, 20]

total_likes, post_status = social_media(post_likes)

print(f"Total Likes: {total_likes}")
print(f"Post Status: {post_status}")
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_analyzer(data, level):
    if not data:
        return 0, "No data available"

    total_rainfall = 0
    for measurement in data:
        total_rainfall += measurement

    average_rainfall = total_rainfall / len(data)

    if average_rainfall >= level:
        rainfall_status = "Adequate Rainfall"
    else:
        rainfall_status = "Inadequate Rainfall"

    return average_rainfall, rainfall_status


rainfall = [65, 78, 72, 68, 77]
rainfall_level = 70

avg_rainfall_adequate, status_adequate = rainfall_analyzer(rainfall,rainfall_level)

print(f"Average Rainfall: {avg_rainfall_adequate:.0f}")
print(f"Rainfall Status: {status_adequate}")

seasonal_rainfall_inadequate = [45, 50, 30, 60, 40]

avg_rainfall_inadequate, status_inadequate = rainfall_analyzer(seasonal_rainfall_inadequate, rainfall_level)

print(f"\nAverage Rainfall: {avg_rainfall_inadequate:.0f}")
print(f"Rainfall Status: {status_inadequate}")
```

Problem 4: Social Media – Duplicate Account Detection


```python
def detect_duplicate_accounts(usernames):

    seen_usernames = set()
    for username in usernames:
        if username in seen_usernames:
            return "Yes"  # Duplicate found
        seen_usernames.add(username)
    return "No"  # No duplicates

user_list_with_duplicates = ["alice123", "bob456", "charlie789", "alice123", "david007"]
duplicate_status_1 = detect_duplicate_accounts(user_list_with_duplicates)
print(f"Duplicate Accounts Found: {duplicate_status_1}")
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_crop_prices(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:  # Premium threshold is ₹2000
            premium_crops.append(price)
    return premium_crops

current_crop_prices = [1500, 2500, 1800, 3200, 1950]
premium_crops_found = filter_crop_prices(current_crop_prices)
print(f"Premium Crops: {premium_crops_found}")
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attended_classes, total_classes):

    if total_classes == 0:
        return 0.0, "No classes conducted"

    attendance_percentage = (attended_classes / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"

    return attendance_percentage, eligibility_status

student_attended_1 = 30
student_total_1 = 40
percentage_1, eligibility_1 = check_exam_eligibility(student_attended_1, student_total_1)
print(f"Attendance Percentage: {percentage_1:.1f}")
print(f"Exam Eligibility: {eligibility_1}")
```

Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units_consumed):
    bill = 0
    usage_classification = ""

    # Calculating bill
    if units_consumed <= 100:
        bill = units_consumed * 3
    elif units_consumed <= 200:
        bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

    # Classify usage level
    if bill < 500:
        usage_classification = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_classification = "Moderate Usage"
    else:
        usage_classification = "High Usage"

    return bill, usage_classification

units_1 = 50
bill_1, usage_1 = calculate_electricity_bill(units_1)
print(f"Units Consumed: {units_1}")
print(f"Total Bill: ₹{bill_1}")
print(f"Usage Classification: {usage_1}")

units_2 = 150
bill_2, usage_2 = calculate_electricity_bill(units_2)
print(f"\nUnits Consumed: {units_2}")
print(f"Total Bill: ₹{bill_2}")
print(f"Usage Classification: {usage_2}")
```
